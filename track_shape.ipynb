```python
from src.utils import openf1
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_sessions = openf1.get_all_sessions()

# Get latest of each track
df_sessions = df_sessions.groupby('circuit_key').last().reset_index()
df_sessions
```

```python

def extract_first_lap(session_key, plot = False):
    df = openf1.fetch_first_lap_locations(session_key, 1)
    circuit_short_name = df_sessions.loc[df_sessions['session_key']==session_key, 'circuit_short_name'].values[0]
    circuit_key = df_sessions.loc[df_sessions['session_key']==session_key, 'circuit_key'].values[0]
    df['circuit_short_name'] = circuit_short_name
    df['circuit_key'] = circuit_key

    if plot:
        plt.scatter(x = df['x'], y = df['y'], linestyle = '-', linewidth = 3)
        plt.title(circuit_short_name)
        
        plt.show()
    
    return df
```

```python
df_tracks = []
for sesh in df_sessions['session_key']:
    df_tracks.append(extract_first_lap(sesh, False))

df_tracks = pd.concat(df_tracks)
```

```python
df_tracks.to_csv('src/assets/tracks.csv')
```
